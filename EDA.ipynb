```python
# ======================================================
# BREAST CANCER PREDICTION - EXPLORATORY DATA ANALYSIS
# ======================================================

# 1. Importarea bibliotecilor necesare
# Pandas si NumPy pentru manipularea datelor
# Matplotlib si Seaborn pentru grafice
# Colab files pentru incarcarea fisierului CSV
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
import io, warnings
warnings.filterwarnings("ignore")

# 2. Incarcarea datasetului de pe Kaggle (fisier CSV)
print("Incarca fisierul CSV de la Kaggle")
uploaded = files.upload()

filename = list(uploaded.keys())[0]
df = pd.read_csv(io.BytesIO(uploaded[filename]))
print("\nDataset incarcat cu succes!")

# Afisam primele 5 randuri pentru a vedea structura datelor
display(df.head())

# 3. Informatii generale despre dataset
# Afisam tipurile de date, numarul de coloane si valorile nenule
print("\nInformatii generale:")
df.info()

# 4. Statistici descriptive pentru coloanele numerice
# Media, mediana, deviatia standard, minim, maxim, quartile
print("\nStatistici descriptive:")
display(df.describe().T)

# 5. Verificam daca exista valori lipsa sau duplicate
print("\nValori lipsa pe coloana:")
print(df.isna().sum())

print("\nNumar de randuri duplicate:", df.duplicated().sum())

# 6. Identificam coloana tinta (diagnosis)
target_col = 'diagnosis' if 'diagnosis' in df.columns else df.columns[-1]
print(f"\nColoana tinta: {target_col}")
print(df[target_col].value_counts())

# 7. Vizualizam distributia claselor tinta (Malign / Benign)
plt.figure(figsize=(5,4))
sns.countplot(x=df[target_col], palette="Set2")
plt.title("Distributia claselor tinta (M/B)")
plt.show()


#8. Eliminam coloanele inutile
drop_cols = [c for c in df.columns if 'id' in c.lower() or 'unnamed' in c.lower()]
if drop_cols:
    df = df.drop(columns=drop_cols)

# Convertim coloana diagnosis din text in valori numerice
if df[target_col].dtype == 'object':
    df[target_col] = df[target_col].map({'M':1, 'B':0})

print("\nDataset curatat")

# 9. Selectarea coloanelor numerice pentru analiza
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
numeric_cols.remove(target_col)

# 10. Analiza univariata - distributia valorilor numerice
# Afisam histograme pentru fiecare variabila numerica
df[numeric_cols].hist(figsize=(16, 12), bins=20, color="#5DADE2")
plt.suptitle("Distributia variabilelor numerice")
plt.show()

# 11. Boxplot - pentru identificarea outlierilor
plt.figure(figsize=(16, 8))
sns.boxplot(data=df[numeric_cols], orient="h", color="#FFCC80")
plt.title("Boxplot")
plt.show()

# 12. Matricea de corelatii
# Corelatia arata relatia liniara dintre variabile
corr = df.corr()
plt.figure(figsize=(14,10))
sns.heatmap(corr, cmap="coolwarm", annot=False)
plt.title("Matricea de corelatii")
plt.show()

# Afisam cele mai corelate variabile cu coloana tinta (diagnosis)
print("\nTop 10 corelatii cu variabila tinta:")
print(corr[target_col].sort_values(ascending=False).head(10))

# 13. Analiza bivariata - comparatia distributiei intre clase
# Grafic KDE pentru primele 4 coloane numerice
for col in numeric_cols[:4]:
    plt.figure(figsize=(6,4))
    sns.kdeplot(data=df, x=col, hue=target_col, fill=True, palette="Set1")
    plt.title(f"Distributia {col} in functie de clasa (0/1)")
    plt.show()

# 14. PCA - reducerea dimensionalitatii si vizualizare 2D
# Folosim StandardScaler pentru normalizarea datelor
# PCA (Principal Component Analysis) transforma datele in doua componente principale
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[numeric_cols])

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
df_pca = pd.DataFrame(X_pca, columns=['PC1','PC2'])
df_pca[target_col] = df[target_col].values

plt.figure(figsize=(8,6))
sns.scatterplot(x='PC1', y='PC2', hue=target_col, data=df_pca, palette="coolwarm")
plt.title("PCA - Vizualizarea claselor in spatiu 2D")
plt.show()

# 15. Sumare finala a informatiilor cheie
print("\nSUMAR FINAL")
print(f"Randuri: {df.shape[0]}, Coloane: {df.shape[1]}")
print("Valori lipsa:", df.isna().sum().sum())
print("Distributie clase:", df[target_col].value_counts().to_dict())
print("Cele mai corelate variabile cu tinta:")
print(corr[target_col].sort_values(ascending=False).head(5))
```
